### Trade-off between accuracy and interpretability

### Variable selection

Our question is now which predictors are associated with the response. We select variables considering AIC, adjusted $R^2$, etc. The problem with this is that 

Least Squares (LS) produces estimates of $\beta_0, \beta_1,\cdots,\beta_p$ by minimising

$$
\begin{align}
RSS = \sum_{i=1}^n\left( y_i-\beta_0 - \sum_{j=1}^p \beta_j x_{i,j}\right)^2
\end{align}
$$

On the other hand, ridge regression minimises

$$
\begin{align}
\sum_{i=1}^n\left( y_i - \beta_0 - \sum_{j=1}^p \beta_j x_{i,j} \right)^2 + \lambda \sum_{j=1}^p \beta_j^2 = RSS + \lambda\sum_{j=1}^p \beta_j^2
\end{align}
$$

where $\lambda\geq0$ is a tuning parameter that controls the trade-off between model fit and the size of the coefficients. When $\lambda=0$ then the penalty term has no effect and the ridge solution is the same as the LS solution. As $\lambda$ gets larger, the penalty term gets larger. The regression coefficients necessarily get smaller in order to minimise the entire function. As $\lambda$ increases the coefficients are shrunk towards 0.

### Bias-variance Trade-off

For a carefullly chosen $\lambda$, ridge regression can significantly outperform LS in terms of estimation and prediction.
